### Activation Functions

```python
# Import required libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate data
x = np.linspace(-5, 5, 200)
```

```python
# activation functions
y_relu = tf.nn.relu(x)
y_sigmoid = tf.sigmoid(x)
y_tanh = tf.nn.tanh(x)
y_softplus = tf.nn.softplus(x)
```

```python
sess = tf.Session()
y_relu, y_sigmoid, y_tanh, y_softplus = sess.run([y_relu, y_sigmoid, y_tanh, y_softplus])
```

```python
# plot activation functions
plt.figure(1, figsize=(8, 6))
plt.subplot(221)
plt.plot(x, y_relu, c='red', label = 'relu')
plt.ylim((-1, 5))
plt.legend(loc='best')

plt.subplot(222)
plt.plot(x, y_sigmoid, c='blue', label='sigmoid')
plt.ylim((-0.2, 1.2))
plt.legend(loc='best')

plt.subplot(223)
plt.plot(x, y_tanh, c='yellow', label='tanh')
plt.ylim((-1.2, 1.2))
plt.legend(loc='best')

plt.subplot(224)
plt.plot(x, y_softplus, c='black', label='softplus')
plt.ylim((-0.2, 6))
plt.legend(loc='best')

plt.show()
```
